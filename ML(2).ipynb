# Create a decision tree classification model for fish.csv dataset. Display confusion matrix,accuracy score, error rate,classification report

```python
#step-1: load dataset
import pandas as pd
df=pd.read_csv("fish.csv")
df.info()
df
```

```python
#step-2: define x and y
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
#step-3: split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#step-4: import model
#step-5: fit or train model, get depth of tree
from sklearn.tree import DecisionTreeClassifier
d_tree=DecisionTreeClassifier(criterion="entropy",max_depth=11)
m=d_tree.fit(x_train,y_train)
print(d_tree.get_depth())
```

```python
#step-6: predict y for x_test
y_pred=d_tree.predict(x_test)
```

```python
#step-7: evaluate the model
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("TN:",cm[0][0])
print("FP:",cm[0][1])
print("FN:",cm[1][0])
print("TP:",cm[1][1])
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

## KNN

```python
import pandas as pd
df=pd.read_csv("fish.csv")
print(df.info())
print(df)

x=df.iloc[:,1:]
y=df.iloc[:,0]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=13)
m=knn.fit(x_train,y_train)

y_pred=knn.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

# Random Forest Classifier

```python
import pandas as pd
df=pd.read_csv("fish.csv")
# print(df.info())

x=df.iloc[:,1:]
y=df.iloc[:,0]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42,max_depth=10)
m=rf.fit(x_train,y_train)

y_pred=rf.predict(x_test)

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

```python
df=pd.read_csv("insurance.csv")
df.head()
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
x=pd.get_dummies(x,drop_first=True)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

m=model.coef_
c=model.intercept_
print("co-efficient: \n",m)
print("Intercept: \n",c)

from sklearn.metrics import mean_squared_error,r2_score
print("MSE:",mean_squared_error(y_test,y_pred))
print("r2Score:",r2_score(y_test,y_pred))
```

# Support Vector Machine

```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
print(df.info())

x=df.iloc[:,0:8]
y=df.iloc[:,-1]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.svm import SVC
svm_model=SVC(kernel="rbf",random_state=42,C=10)
#C: regulazitaion parameter of C will increase margin distance & reduce error 
#it may go in overfitting
#Low value of C will decrease marginal distance & reduce accuracy
#kernel will convert dataset from higher dimension to make it separable
#rbf: radial bias function, most powerfull & commonly used
svm_model.fit(x_train,y_train)

y_pred=svm_model.predict(x_test)

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("TN:",cm[0][0])
print("FP:",cm[0][1])
print("FN:",cm[1][0])
print("TP:",cm[1][1])
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

```python
# Data Preprocessing
import pandas as pd
df=pd.read_csv("Employee_Attrition.csv")
df.info()
df=df.drop("Employee_ID",axis=1)
df=pd.get_dummies(df,drop_first=True,dtype="int")
print(df)
```

```python
#KNN
x=df.drop("Attrition",axis=1)
y=df["Attrition"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=42)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
acc_list=[]
for k in range(1,26):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    acc_list.append(acc)
    print(f"k={k},Accuracy={acc}")

max_acc=max(acc_list)
best_k=(acc_list.index(max_acc))+1
print("best k:",best_k)
knn=KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train,y_train)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=150,random_state=42,
                          max_depth=12)
m=rf.fit(x_train,y_train)

y_pred1=rf.predict(x_test)
acc=accuracy_score(y_test,y_pred1)
print("accuracy of RF:",acc)
cm=confusion_matrix(y_test,y_pred1)
print("cm for RF: \n",cm)
```

```python
#Prediction Task
import numpy as np
x_t=np.array([30,45000,5,3,10,1,0,0,1,0,1,0,0]).reshape(1,-1)
print(x_t)
print("prediction for knn model:",knn.predict(x_t))
print("prediction for Rf model:",rf.predict(x_t))
```

```python
# SVM
from sklearn.svm import SVC
svm=SVC(kernel="rbf",random_state=42,C=10)
svm.fit(x_train,y_train)
y_pred2=svm.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred2)
print(cm)
print("TN:",cm[0][0])
print("FP:",cm[0][1])
print("FN:",cm[1][0])
print("TP:",cm[1][1])
print("accuracy:",accuracy_score(y_test,y_pred2))
print("classification report: \n",classification_report(y_test,y_pred2))
```
