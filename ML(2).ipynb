# Create a decision tree classification model for fish.csv dataset. Display confusion matrix,accuracy score, error rate,classification report

```python
#step-1: load dataset
import pandas as pd
df=pd.read_csv("fish.csv")
df.info()
df
```

```python
#step-2: define x and y
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
#step-3: split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#step-4: import model
#step-5: fit or train model, get depth of tree
from sklearn.tree import DecisionTreeClassifier
d_tree=DecisionTreeClassifier(criterion="entropy",max_depth=11)
m=d_tree.fit(x_train,y_train)
print(d_tree.get_depth())
```

```python
#step-6: predict y for x_test
y_pred=d_tree.predict(x_test)
```

```python
#step-7: evaluate the model
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("TN:",cm[0][0])
print("FP:",cm[0][1])
print("FN:",cm[1][0])
print("TP:",cm[1][1])
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

## KNN

```python
import pandas as pd
df=pd.read_csv("fish.csv")
print(df.info())
print(df)

x=df.iloc[:,1:]
y=df.iloc[:,0]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=13)
m=knn.fit(x_train,y_train)

y_pred=knn.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

## Random Forest Classifier

```python
import pandas as pd
df=pd.read_csv("fish.csv")
# print(df.info())

x=df.iloc[:,1:]
y=df.iloc[:,0]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42,max_depth=10)
m=rf.fit(x_train,y_train)

y_pred=rf.predict(x_test)

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

```python
import pandas as pd
df=pd.read_csv("iris.csv")
df.info()
df

x=df.iloc[:,1:-1]
y=df.iloc[:,-1]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42,max_depth=10)
m=rf.fit(x_train,y_train)

y_pred=rf.predict(x_test)

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report: \n",classification_report(y_test,y_pred))
```

```python
df=pd.read_csv("insurance.csv")
df.head()
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
x=pd.get_dummies(x,drop_first=True)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

m=model.coef_
c=model.intercept_
print("co-efficient: \n",m)
print("Intercept: \n",c)

from sklearn.metrics import mean_squared_error,r2_score
print("MSE:",mean_squared_error(y_test,y_pred))
print("r2Score:",r2_score(y_test,y_pred))
```
