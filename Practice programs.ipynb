```python
# Write a python program to predict reservation using svm and random forest algorithm for Hotel Reservation.csv
# dataset(drop null values if req, convert categorical to numeric if req).
# Split dataset in training and testing(80%-20%, random_state=42)

#Random forest:
# train random forest model for values of n_estimators 50 to 100
# random_state=42, criterion="entropy"
# find best n_estimators based on highest accuracy
# train the final random forest model using optimal n_estimators

#svm
# train svc model using random_state=42,C=10 and 
# find accuracy_score and confusion matrix
```

```python
import pandas as pd
df=pd.read_csv("Hotel Reservations.csv")
print(df.info())
df=df.drop("Booking_ID",axis=1)
x=df.drop("booking_status",axis=1)
x=pd.get_dummies(x,drop_first=True,dtype="int")
y=df["booking_status"]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=40)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
acc_list=[]
for e in range(50,101):
    rf=RandomForestClassifier(n_estimators=e,random_state=42,criterion="entropy")
    rf.fit(x_train,y_train)
    y_pred=rf.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    acc_list.append(acc)
    print(f"estimators={e},Accuracy={acc}")
max_acc=max(acc_list)

best_e=(acc_list.index(max_acc))+1
print("best e:",best_e)
rf_final=RandomForestClassifier(n_estimators=best_e,random_state=42,criterion="entropy")
rf_final.fit(x_train,y_train)
y_pred1=rf_final.predict(x_test)
acc=accuracy_score(y_test,y_pred1)
print("accuracy of Rf:",acc)
cm=confusion_matrix(y_test,y_pred1)
print("cm for RF: \n",cm)

#svc
from sklearn.svm import SVC
s=SVC(C=10,random_state=42)
s.fit(x_train,y_train)
y_pred1=s.predict(x_test)
acc=accuracy_score(y_test,y_pred1)
print("accuracy of svc:",acc)
cm=confusion_matrix(y_test,y_pred1)
print("cm for svc: \n",cm)
```
