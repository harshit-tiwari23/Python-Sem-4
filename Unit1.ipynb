## with user defined function

```python
import pandas as pd
df=pd.DataFrame({"maths":[80,85,90],"sci":[80,75,78],"eng":[70,72,75]})
def grade(total):
    if total>270:
        return "A"
    elif total>230:
        return "B"
    else:
        return "C"

df["total"]=df.apply(lambda x:x.sum(),axis=1)
df["grade"]=df["total"].apply(grade)
print(df)
```

```python
import pandas as pd
df=pd.DataFrame({"maths":[80,85,90],"sci":[80,75,78],"eng":[70,72,75]})
print(df)
def bonus(x,b):
    return x+b

# add 5 marks in all
df1=df.apply(bonus,axis=1,args=(5,))
print(df1)
```

```python
df=pd.DataFrame({"maths":[80,85,90],"sci":[80,75,78],"eng":[70,72,75]})
print(df)
df.apply(lambda x:[x.sum(),x.mean()],result_type="expand",axis=1)
```

```python
df=pd.DataFrame({"maths":[80,85,90],"sci":[80,75,78],"eng":[70,72,75]})
print(df)
df.apply(lambda x:x.sum()-x,result_type="broadcast",axis=1)
```

# concat()

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"B":[7,8]})
res=pd.concat([df1,df2])
res
```

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"B":[7,8]})
res=pd.concat([df1,df2],axis=1)
res
```

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"C":[5,6],"D":[7,8]})
res=pd.concat([df1,df2])
res
```

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"C":[7,8]})
res=pd.concat([df1,df2])
res
```

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"C":[5,6],"D":[7,8]})
res=pd.concat([df1,df2],ignore_index=True)
res
```

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"C":[5,6],"D":[7,8]})
res=pd.concat([df1,df2],join="outer")      # join=inner/outer
res
```

```python
df1=pd.DataFrame({"X":[10,20],"B":[3,4]},index=[12,9])
df2=pd.DataFrame({"A":[5,6,7],"D":[7,2,8]},index=[11,6,13])
res=pd.concat([df1,df2],sort=True,axis=1)      # sort=True/False
res
```

# merge()

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                       "dept_id":[10,20,30,40]})
dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
pd.merge(employees,dept)
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                       "dept_id":[10,20,30,40]})
dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
pd.merge(employees,dept,on="dept_id",how="outer")       # how=inner/outer/left/right
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                       "dept_id":[10,20,30,40]})
dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
pd.merge(employees,dept,on="dept_id",how="right")       # how=inner/outer/left/right
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                       "dept_id":[10,20,30,40]})
dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
pd.merge(employees,dept,on="dept_id",how="left")       # how=inner/outer/left/right
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                        "dept_id":[10,20,30,40]})
dept=pd.DataFrame({"d_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
pd.merge(employees,dept,left_on="dept_id",right_on="d_id")
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                        "dept_id":[10,20,30,40]})
dept=pd.DataFrame({"d_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
pd.merge(employees,dept,left_on="dept_id",right_on="d_id").drop("d_id",axis=1)
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                        "dept_id":[10,20,30,40]})
dept=pd.DataFrame({"d_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
pd.merge(employees,dept,left_index=True,right_index=True)
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                        "dept_id":[10,20,30,40],
                       "updated":["march","april","may","june"]})
dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"],
                  "updated":["march","april","may"]})
pd.merge(employees,dept,on="dept_id",suffixes=("_emp","_dept"))
```

```python
employees=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],
                        "dept_id":[10,20,30,40],
                       "updated":["march","april","may","june"]})
dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"],
                  "updated":["march","april","may"]})
pd.merge(employees,dept,on="dept_id",right_index=True,left_index=True)
```
